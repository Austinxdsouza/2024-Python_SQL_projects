```python
# import datetime as dt
import requests
import pandas as pd
from pytrends.request import TrendReq
```

CONNECT TO GOOGLE

```python
pytrends = TrendReq(hl='en-US', tz=360)
```

```python
kw_list = ["recession"]
```

```python
# last year of data
pytrends.build_payload(kw_list,
                        cat=0,
                        timeframe='today 12-m',
                        geo='',
                        gprop=''
                      )

iot=pytrends.interest_over_time()
iot.plot()
```
