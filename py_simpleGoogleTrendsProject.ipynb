```python
# import datetime as dt
import requests
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
```

CONNECT TO GOOGLE

```python
pytrends = TrendReq(hl='en-US', tz=360)
```

```python
kw_list = ["football","Rugby","Tennis"]
```

```python
# last year of data
pytrends.build_payload(kw_list,
                        cat=0,
                        timeframe='today 3-m',
                        geo='',
                        gprop=''
                      )

iot=pytrends.interest_over_time()
#iot.plot(subplots=True,figsize=(15,12))
Grouped_iot=iot.groupby(pd.Grouper(freq='M'))
Grouped_iot.plot(kind="bar")
plt.show()
```
